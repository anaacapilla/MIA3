In the first place we are going to load the data and preprocess it. This includes removing the NaN values, analyse it and split the dataset.

```
import seaborn as sns
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```
dataset = sns.load_dataset("penguins")
```

```
dataset.head()
```

```
dataset.dropna(inplace = True)
```

```
dataset.head()
```

```
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1:]
```

```
X_dummies = pd.get_dummies(X)
```

```
X_train, X_test, y_train, y_test = train_test_split(X_dummies, y, test_size = 0.2)
```

Now we are going to import the model packages needed. In this case, we selected the **Support Vector Machine** (SVM) for doing the exercise. 

FALTA EXPLICACIN DE CMO FUNCIONA EL SVM


```
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.svm import SVC
```

```
y_train = pd.Series(y_train["sex"]).ravel()
y_test = pd.Series(y_test["sex"]).ravel()
```

```
model = SVC(kernel = 'linear')
model.fit(X_train, y_train)
# IMP AQUI HIPERPARAMETROS: ELKERNEL Y LA GAMMA Y LA ALPHA
```

```
model.score(X_train, y_train)
```

```
model.score(X_test, y_test)
```
